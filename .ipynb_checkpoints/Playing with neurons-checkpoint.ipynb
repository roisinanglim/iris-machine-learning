### Import Packages

```python
import matplotlib.pyplot as plt
import numpy as np
import keras as kr
```

```python
# Resize plots
plt.rcParams['figure.figsize'] = (10, 10)
```

#### Generate data and create arrays

```python
#Creates numpy array to generate 2000 data points between 10 & -10
x = np.linspace(-10.0, 10.0, 2000)

# Generate linear array
y_l = (3.0 * x) + 5.0
#Generate polynomial array
y_p = (x * x) - 41.0
#Generate Sine array
y_s = 10.0 * np.sin(2.0 * x)
```

```python
# Plot grapg using latek to label graph
plt.plot(x, y_l, label='$3x + 5$')
plt.plot(x, y_p, label='$x^2 - 41$')
plt.plot(x, y_s, label='$10 \sin(2x)$')

plt.legend()
```

```python
#### Create neural network trained to take inu values and mimick arrays
```

```python
#Relabel arrays from above & predict one array at a time
#y = y_l
#y = y_p
y = y_s

# Create a new neural network.
m = kr.models.Sequential()

# Add neurons.
#m.add(kr.layers.Dense(1, input_dim=1, activation="linear"))

# Add 3 alayers of neurons with one linear output
m.add(kr.layers.Dense(10, input_dim=1, activation="tanh"))
m.add(kr.layers.Dense(10, activation="linear"))
m.add(kr.layers.Dense(1, activation='linear'))

# Compile the model.
m.compile(loss="mean_squared_error", optimizer="adam")

# Train the model on the 2000 datapoints.- use this dataset 15 times. Eachtime take 10 values
m.fit(x, y, epochs=15, batch_size=10)

# Run each x value through the neural network.
p = m.predict(x)

# Plot the values.Plot predictions as red value and data as black dotted line
plt.plot(x, p, 'r-', label='Prediction')
plt.plot(x, y, 'k:', label='Original')
plt.legend()
```
